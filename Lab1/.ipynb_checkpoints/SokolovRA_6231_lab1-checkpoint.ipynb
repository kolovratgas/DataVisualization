```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('data/data_TV.csv')
```

```python
df = df.dropna()
```

```python
df['year'] = df[df.columns[0]].apply(lambda x: datetime.datetime.strptime(str(x), '%Y-%m-%d').date().year)
```

```python
df.head(5)
```

## График 1 - соотношение популярности к количеству оценок пользователей

```python
sns.relplot(data=df, x='vote_count', y='popularity')
```

## График 2 - Рейтинг стран с наивысшими оценками сериалов

```python
df_good = df[df['origin_country'].apply(lambda x: len(x)) == 2] 
```

```python
df_good = df_good[df_good['vote_average'] > 8]
```

```python
plt.figure(figsize = (12,8))
plt.xticks(rotation = 90)
sns.boxplot(data = df_good, y='vote_average', x='origin_country')
```

## График 3 - Рейтинг стран (Америка и Япония) по количеству выпущенных сериалов по годам c 2020 года

```python
df_count = df.groupby(['origin_country', 'year']).agg('count')['name'].reset_index().sort_values('name')
```

```python
df_count = df_count[df_count['origin_country'].str.contains('US|JP') * df_count['origin_country'].apply(lambda x: len(x) == 2)]
```

```python
df_count = df_count[df_count['year'] >= 2020]
```

```python
plt.figure(figsize = (25,15))
sns.catplot(data=df_count, x="origin_country", y="name", col="year", kind="bar")
```

## График 4 - Количество выпущенных сериалов по годам в 21 веке

```python
df_year = df.groupby('year').agg('count')['name'].reset_index().sort_values('name', ascending=False)
```

```python
df_year = df_year[df_year['year'] >= 2000]
```

```python
plt.figure(figsize = (12,8))
plt.xticks(rotation = 90)
sns.barplot(data=df_year, x='year', y='name')
```
